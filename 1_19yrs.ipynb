```python


import pandas as pd


df = pd.read_excel('1-19yrs.xlsx')

# Print the first few rows of the dataframe
print(df.head())
```

```python


from IPython.display import display, HTML

# Convert the DataFrame to an HTML table
html_table = df.to_html()

# Display the HTML table
display(HTML(html_table))
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'donor_age' and 'structure_name' are columns in your DataFrame

plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.scatterplot(x='donor_age', y='structure_name', data=df, hue='structure_name', palette='bright')
plt.title('Donor Age vs Structure Name')
plt.xlabel('Donor Age')
plt.ylabel('Structure Name')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Structure Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.countplot(x='donor_age', hue='structure_name', data=df)
plt.title('Donor Age vs Structure Name')
plt.xlabel('Donor Age')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Structure Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='structure_name', y='donor_age', data=df)
plt.title('Donor Age Distribution by Structure Name')
plt.xlabel('Structure Name')
plt.ylabel('Donor Age')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'structure_name' and 'donor_age' are columns in your DataFrame

# Group the data by 'structure_name' and count the occurrences of each age
structure_age_counts = df.groupby(['structure_name', 'donor_age'])['donor_age'].count().unstack()

# Create the stacked bar chart
structure_age_counts.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Donor Age Distribution by Structure Name (Stacked Bar Chart)')
plt.xlabel('Structure Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Donor Age')
plt.tight_layout()
plt.show()
```
